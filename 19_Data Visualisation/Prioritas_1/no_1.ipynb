```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('electronic_shop.csv')
```

```python
### No.1 - a
transaction_by_date = df.groupby(['transaction_date']).sum(numeric_only=True)
# transaction_by_date
transaction_dates = transaction_by_date.index.get_level_values(0).to_list()
amount_averages = transaction_by_date['transaction_amount']

plt.figure(figsize=(15,7))
plt.title("Transaction Averages by Date")
plt.xlabel("Date")
plt.ylabel("Average")

plt.plot(transaction_dates, amount_averages)
for i in range(len(transaction_dates)):
    plt.text(transaction_dates[i], amount_averages[i], f'{amount_averages[i]:.2f}', ha='center', va='bottom')
plt.show()
```

```python
### No.1 - b
customer_per_day = df.groupby('transaction_date')['customer_name'].nunique()
dates = customer_per_day.index.get_level_values(0).to_list()

plt.xlabel("Dates")
plt.ylabel("Customer Amount")
plt.bar(dates, customer_per_day)

plt.show()
```

```python
### No.1 - c
transaction_by_date = df.groupby(['transaction_date']).sum(numeric_only=True)
transaction_dates = transaction_by_date.index.get_level_values(0).to_list()
amount_averages = transaction_by_date['transaction_amount']

plt.figure(1, figsize=(15,7))
plt.scatter(transaction_dates, amount_averages)

plt.figure(2, figsize=(15,7))
plt.plot(transaction_dates, amount_averages)

plt.show()
```
