```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np

df = pd.read_csv('survey.csv')
```

```python
### No.2 - a
responden_by_age = df.groupby(['age']).count()
ages = responden_by_age.index.get_level_values(0).tolist()
responden_count = responden_by_age['name']

plt.pie(
    responden_count,
    labels=ages,
    autopct="%.0f%%"
)
plt.show()
```

```python
### No.2 - b

df.replace('-', np.nan, inplace=True)
data_cleaned = df.dropna()

framework_on_frontend = data_cleaned.groupby(['favorite_front_end']).count()
framework = framework_on_frontend.index.get_level_values(0).to_list()
framework_count = framework_on_frontend['name']

plt.xlabel("Framework")
plt.xticks(rotation=80)
plt.bar(framework, framework_count)

plt.show()
```

```python
### No.2 - c 
framework_on_backend = df.groupby(['favorite_back_end']).count()
framework = framework_on_backend.index.get_level_values(0).to_list()
framework_count = framework_on_backend['name']

plt.xlabel("Framework")
plt.xticks(rotation=80)
plt.bar(framework, framework_count)

plt.show()
```

```python
### No.2 - e
survey_by_salary = df.groupby(['salary_per_year_in_USD']).count()
salary_range = survey_by_salary.index.get_level_values(0).to_list()
count = survey_by_salary['name']

plt.pie(
    count,
    labels=salary_range,
    autopct="%.0f%%"
)

plt.show()
```
