```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bakery.csv')
```

```python
### No.2 - a
transaction_by_date = df.groupby(['transaction_date']).sum(numeric_only=True)
transaction_dates = transaction_by_date.index.get_level_values(0).to_list()
amount_sum = transaction_by_date['transaction_amount']

plt.xlabel("Dates")
plt.ylabel("Transaction Amount")
plt.xticks(rotation=80)
plt.bar(transaction_dates, amount_sum)

plt.show()
```

```python
### No.2 - b
transaction_by_category = df.groupby(['item_name']).count()
item_category = transaction_by_category.index.get_level_values(0).to_list()
transaction_count = transaction_by_category['transaction_id']


plt.pie(
    transaction_count,
    labels=item_category,
    autopct="%.0f%%"
)
plt.show()
```

```python
### No.2 - c
transaction_by_date = df.groupby(['transaction_date']).mean(numeric_only=True)
transaction_dates = transaction_by_date.index.get_level_values(0).to_list()
rating_mean = transaction_by_date['customer_rating']

plt.figure(figsize=(15,7))
plt.title("Rata-Rata Rating Customer / Hari")
plt.xlabel("Dates")
plt.ylabel("Mean")
plt.plot(transaction_dates, rating_mean)
plt.show()
```
