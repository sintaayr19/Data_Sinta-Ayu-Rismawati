```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('e-commerce.json')
```

```python
### No.1 - a
transaction_by_date = df.groupby(['transaction_date']).sum(numeric_only=True)
dates = transaction_by_date.index.get_level_values(0).to_list()
transaction_sum = transaction_by_date['transaction_amount']

plt.figure(figsize=(15,7))
plt.xlabel("Dates")
plt.ylabel("Transaction Amount")
plt.bar(dates, transaction_sum)
plt.xticks(rotation=80)
plt.show()
```

```python
### No.1 - b 
transaction_by_category = df.groupby('item_category').count()
item_category = transaction_by_category.index.get_level_values(0).to_list()
amount_sold = transaction_by_category['transaction_id']

plt.figure(figsize=(15,7))
plt.xlabel("Category")
plt.ylabel("Amount Sold")
plt.bar(item_category, amount_sold)
plt.show()
```

```python
### No.1 - c 
transaction_e_wallet = df.loc[df['payment_method'] == 'e wallet'].sum(numeric_only=True)
transaction_e_wallet['transaction_amount']

result_df = pd.DataFrame({'payment_method': 'e wallet', 'transaction_total': [transaction_e_wallet['transaction_amount']]})
result_df
```

```python
### No.1 - d 
transaction_by_date = df.groupby(['transaction_date']).sum(numeric_only=True)
dates = transaction_by_date.index.get_level_values(0).to_list()
transaction_amount = transaction_by_date['transaction_amount']

plt.boxplot(transaction_amount)
plt.show()
```
